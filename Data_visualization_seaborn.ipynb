```python
#import libraries
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
print ('libraries are sucessfull imported')
```

```python
df=pd.read_csv('amenity_pop.csv')
print(df)
```

```python
plt.figure(figsize=(7,4))
sns.histplot(data=df,bins=6, x='population', color='steelblue',
             kde=True #smooth line
             )
plt.title('population distribution')
plt.xlabel('population')
plt.ylabel('count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(data=df,x='amenity', palette='Set2')
plt.title('Number of Facilities per Amenity Type')
plt.xlabel('Amenity Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(data=df,x='amenity',hue='status', palette='Set2')
plt.title('Number of Facilities per Amenity Type')
plt.xlabel('Amenity Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
#Bar plot
plt.figure(figsize=(7,4))
sns.barplot(data=df,x='district',y='population',palette='Blues_d',errorbar=None)
plt.title('Average Population per Distict')
plt.xlabel('District')
plt.ylabel('Average Population')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=df,x='amenity',y='satisfaction',hue='status',palette='Set2',errorbar=None)
plt.title(('Average Satisfaction per Amenity Type'))
plt.ylabel('Average Satisfaction')
plt.xlabel('Amenity')
plt.tight_layout()
plt.show()
```

```python
#Box plot
plt.figure(figsize=(7,4))
sns.boxplot(data=df,x='amenity', hue='amenity',y='population', palette='pastel')
plt.title('Population Range per Amenity Type')
plt.xlabel('Amenity')
plt.ylabel('Population')
plt.tight_layout()
plt.show()
```

```python
plt.figsize=(7,4)
sns.scatterplot(data=df, x='population',y='satisfaction', color='black', s=80)
plt.title('Population vs Satisfaction')
plt.xlabel('Population')
plt.xlabel('Satisfaction Score(1-5)')
plt.tight_layout()
plt.show()
```

```python
#Create a simple year vs satisfaction example
year_data = pd.DataFrame({
    'year': [2018, 2019, 2020, 2021, 2022, 2023],
    'satisfaction': [3.4, 3.6, 3.5, 3.8, 4.0, 4.2]
})

plt.figure(figsize=(7,4))

sns.lineplot(data=year_data, x='year', y='satisfaction', marker='o', color='teal', linewidth=2.5)

plt.title("Average Satisfaction Over the Year")
plt.xlabel("Year")
plt.ylabel("Satisfaction Score(1-5)")

plt.tight_layout()
plt.show()

print("marker='o' adds a dot at each data point")
```
